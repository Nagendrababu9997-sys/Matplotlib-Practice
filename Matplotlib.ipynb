**Multiple** **Bars**

```python
import matplotlib.pyplot as plt
import numpy as np

products = ["Laptop", "Mobile", "Tablet"]

sales_2024 = [120, 200, 80]

sales_2025 = [150, 220, 90]

x = np.arange(len(products))

plt.bar(x - 0.2, sales_2024, width=0.4, label="2024")
plt.bar(x + 0.2, sales_2025, width=0.4, label="2025")

plt.xticks(x, products)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

months=["Jan","Feb","Mar","Apr"]

sales=[120,180,150,220]

plt.figure(figsize=(8,5))

plt.plot(
months,
sales,
color="green",
marker="o"
)

plt.xlim("Jan","Apr")

plt.ylim(100,250)

plt.xticks(rotation=45)

plt.yticks([100,150,200,250])

plt.title("Monthly Sales")

plt.xlabel("Months")

plt.ylabel("Sales")

plt.grid(True)

plt.style.use("ggplot")

plt.savefig("sales_report.png")

plt.show()
```
